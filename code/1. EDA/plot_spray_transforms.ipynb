```python
def spray_transform(num_days_array, decay_width=10, decay_end=20):
    midpoint = decay_end - decay_width/2
    scale = 10/decay_width
    
    trans_array = -1/(1+np.exp(-scale*(num_days_array-midpoint)))+1
    
    return trans_array
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('dark')

%matplotlib inline
```

```python
x = np.linspace(0,2000, 100)
y = np.log(x)

fig = plt.figure(figsize=(6,5))
ax = fig.gca()

plt.plot(x,y)
ax.set_title('Log days', fontsize='16')
ax.set_ylabel('Log days', fontsize='12')
ax.set_xlabel('Number of days since last spray', fontsize='12')
```

```python
x = np.linspace(0,30, 100)
y = spray_transform(x)

fig = plt.figure(figsize=(6,5))
ax = fig.gca()

plt.plot(x,y)
ax.set_title('Spray Effectiveness', fontsize='16')
ax.set_ylabel('Spray Effectiveness', fontsize='12')
ax.set_xlabel('Number of days since last spray', fontsize='12')
```
