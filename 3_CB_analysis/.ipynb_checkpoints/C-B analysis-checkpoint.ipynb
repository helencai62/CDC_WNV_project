```python
import pandas as pd
import numpy as np

from sklearn.externals import joblib
from sklearn.metrics import confusion_matrix

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load model
gs_bal = joblib.load('../2_Models/lr_bal_gs.pkl')

# load test data
lr_test_data = pd.read_csv('../datasets/lr_test_data.csv', index_col=0)
X_test = lr_test_data.drop(columns=['WnvPresent'])
y_test = lr_test_data['WnvPresent']
```

#### Calculating the costs and benefits per square mile

```python
# benefit = medical benefits saved by not having to treat patients

pp_medical_costs = 99.336 # in $000s, source: https://wwwnc.cdc.gov/eid/article/16/3/09-0667_article
pop_per_sq_mile = 11923 # source: https://www.opendatanetwork.com/entity/1600000US1714000/Chicago_IL/geographic.population.density?year=2016

sm_medical_cost = pp_medical_costs*pop_per_sq_mile
```

```python
sm_medical_cost
```

```python
# cost = cost of spray per sq mile

single_spray_cost = 47.122 # in $000s, source: https://wwwnc.cdc.gov/eid/article/16/3/09-0667_article

# spray from 25 July 2013 looks very square, based on Tableau graph
# use this to calculate number of sprays needed to cover a square area

spray = pd.read_csv('../datasets/spray.csv')
spray_sq = spray[spray['Date']=='2013-07-25']

lat_range = spray_sq['Latitude'].max() - spray_sq['Latitude'].min()
long_range = spray_sq['Longitude'].max() - spray_sq['Longitude'].min()
num_sprays = spray.shape[0]

# At 40 degress north, one degree of latitude ~ 70 miles
# At 40 degress north, one degree of longitude ~ 50 miles
# source: http://www.longitudestore.com/how-big-is-one-gps-degree.html
miles_across = long_range*50
miles_high = lat_range*70
sq_miles = miles_high*miles_across

sprays_per_sq_mile = num_sprays/sq_miles
sm_spray_cost = sprays_per_sq_mile*single_spray_cost
```

```python
thresholds = np.linspace(0.01,1,100)

diseased_perc = []
sprayed_perc = []

for i, threshold in enumerate(thresholds):
    pred = gs_bal.predict_proba(X_test)
    y_test_hat = []
    for obs in range(len(pred)):
        if pred[obs][1] < threshold:
            y_test_hat.append(0)
        else:
            y_test_hat.append(1)

    conf_mat = confusion_matrix(y_test, y_test_hat)
    conf_mat = conf_mat/(conf_mat[0,0]+conf_mat[0,1]+conf_mat[1,0]+conf_mat[1,1])
    
    diseased_perc.append(conf_mat[1,0]) # FN
    sprayed_perc.append(conf_mat[1,1]+conf_mat[0,1]) # FP + TP
    
```

```python
total_cost = np.array(sprayed_perc)*sm_spray_cost + np.array(diseased_perc)*sm_medical_cost

fig = plt.figure(figsize=(16, 8))
ax = plt.gca()
plt.plot(thresholds, total_cost/1000)
ax.set_title('How total cost varies by model threshold', fontsize=20)
ax.set_xlabel('Model threshold', fontsize=16)
ax.set_ylabel('Cost ($m)', fontsize=16)
ax.set_ylim((0,200));
# threshold close to 0 means we're spraying more locations, as more locations get counted as having WNV
```

```python
min_cost = min(total_cost)/1000
min_loc = sprayed_perc[np.argmin(total_cost)]*100

fig = plt.figure(figsize=(12, 8))
ax = plt.gca()
plt.plot(np.array(sprayed_perc)*100, total_cost/1000)
ax.set_title('How total cost varies by percentage of locations sprayed', fontsize=20)
ax.set_xlabel('Locations sprayed (%)', fontsize=16)
ax.set_ylabel('Cost ($m)', fontsize=16)
ax.set_ylim((0,200))
ax.axhline(min_cost, linestyle='dashed', c='red')
ax.axvline(min_loc, linestyle='dashed', c='red')
ax.text(x=75, y=min_cost+5, s='$'+str(round(min_cost,1))+'m', fontsize=20, color='r')
ax.text(x=min_loc+1, y=25, s=str(round(min_loc, 2))+'%', fontsize=20, color='r');
```
